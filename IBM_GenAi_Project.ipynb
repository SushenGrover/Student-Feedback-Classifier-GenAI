Classification using Flan T5 model for each student review

```python
!pip install transformers torch pandas | tail -n 1

import pandas as pd
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

df = pd.read_csv('fabricated_data.csv')

categories = [
    "Academics", "Facilities", "Administration", "Faculty",
    "Placements & Internships", "Hostel & Accommodation",
    "Technical Infrastructure", "Extracurricular Activities",
    "Health & Wellness", "Transport & Accessibility"
]

# load the model and tokenizer
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def classify_review(text, categories):
    prompt = (
        f"Classify the following student feedback into one of these categories: {', '.join(categories)}.\n"
        f"Feedback: {text}\n"
        "Category:"
    )
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)
    outputs = model.generate(**inputs, max_new_tokens=10)
    prediction = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # Clean prediction to match category list
    for cat in categories:
        if cat.lower() in prediction.lower():
            return cat
    # If no match, return raw prediction
    return prediction.strip()

# Predict categories
df['predicted_category'] = df['feedback'].apply(lambda x: classify_review(x, categories))

# Save to new CSV
df.to_csv('predicted_dataset.csv', index=False)
print("Predictions saved to predicted_dataset.csv")
```

Comparing the results with the original answers

```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix

# assuming df is already loaded with columns 'category' and 'predicted_category'
categories = [
    "Academics", "Facilities", "Administration", "Faculty",
    "Placements & Internships", "Hostel & Accommodation",
    "Technical Infrastructure", "Extracurricular Activities",
    "Health & Wellness", "Transport & Accessibility"
]

# generate classification report as a dictionary
report_dict = classification_report(
    df['category'], df['predicted_category'], labels=categories, output_dict=True
)

# convert classification report to DataFrame for pretty display
report_df = pd.DataFrame(report_dict).transpose()
report_df = report_df.round(2)  # Round for neatness

#classification report
print("Classification Report:")
display(report_df)

#confusion matrix as DataFrame
cm = confusion_matrix(df['category'], df['predicted_category'], labels=categories)
cm_df = pd.DataFrame(cm, index=categories, columns=categories)

print("Confusion Matrix:")
display(cm_df)
```

Comparison with Visulizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# Assuming df is already loaded with columns 'category' and 'predicted_category'
categories = [
    "Academics", "Facilities", "Administration", "Faculty",
    "Placements & Internships", "Hostel & Accommodation",
    "Technical Infrastructure", "Extracurricular Activities",
    "Health & Wellness", "Transport & Accessibility"
]

# Classification Report Visualization (Bar Plot)
report = classification_report(
    df['category'], df['predicted_category'], labels=categories, output_dict=True
)
report_df = pd.DataFrame(report).transpose().loc[categories, ['precision', 'recall', 'f1-score']]
report_df = report_df.round(2)

plt.figure(figsize=(14, 7))
report_df.plot(kind='bar', figsize=(14, 7), width=0.8, colormap='viridis')
plt.title('Classification Report Metrics by Category', fontsize=20, pad=20)
plt.ylabel('Score', fontsize=15)
plt.xlabel('Category', fontsize=15)
plt.xticks(rotation=30, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0, 1)
plt.legend(fontsize=13, loc='lower right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Confusion Matrix Visualization (Heatmap)
cm = confusion_matrix(df['category'], df['predicted_category'], labels=categories)
cm_df = pd.DataFrame(cm, index=categories, columns=categories)

plt.figure(figsize=(12, 10))
sns.set(font_scale=1.2)
ax = sns.heatmap(
    cm_df,
    annot=True,
    fmt='d',
    cmap='YlGnBu',
    linewidths=0.5,
    linecolor='gray',
    cbar_kws={"shrink": .8}
)
plt.title('Confusion Matrix', fontsize=20, pad=20)
plt.ylabel('Actual Category', fontsize=15)
plt.xlabel('Predicted Category', fontsize=15)
plt.xticks(rotation=30, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.tight_layout()
plt.show()
```
